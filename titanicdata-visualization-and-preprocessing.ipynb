```python
from matplotlib import pyplot as plt
import seaborn as sns
```

# Count Plot

```python
sns.countplot(x = data['Pclass'], data = data)
plt.title('Count Plot')
plt.xlabel('Pclass')
plt.ylabel('Count')
```

# Histogram

```python
sns.histplot(data['Age'],bins=10)
plt.title('Hist plot')
plt.xlabel('Age')
plt.ylabel('Count')
```

# Box Plot

```python
sns.boxplot(data=data,x = data['Pclass'], y=data['Fare'],hue=data['Sex'])
plt.title('Box plot')
#plt.xlabel('Age')
plt.ylabel('fare')
```

# Bar Plot

```python
sns.barplot(x=data['Fare'], y=data['Age'], data=data, ci=None)
plt.title('Bar Plot')
plt.xlabel('Fare')
plt.ylabel('Age')
```

# PiPlot

```python
education_counts = data['Pclass'].value_counts()
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
#plt.title('PiPlot')
plt.title('Distribution of PClass')
plt.axis('equal')
```

# Scatter Plot

```python
sns.scatterplot(x='Age', y='Fare', data=data)
plt.title('Scatter Plot of Age vs. Income')
```

# Violin Plot

```python
sns.violinplot(x='Sex', y='Age', data=data)
plt.title('Age Distribution by Gender')
```

# Pair Plot

```python
plt.subplot(3, 3, 4)
sns.pairplot(data)
plt.title('Pairplot of Numerical Columns')
```

# HeatMap

```python
sns.heatmap(data[['Age', 'Fare']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Columns')
```

# Line Plot

```python
sns.lineplot(x='Age', y='Fare', data=data, marker='o', linestyle='-')
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Age'])
#sns.histplot(data['Age'], kde=True, stat='density', color='skyblue', bins=30, element='step')
plt.title('Probability Density Function (PDF)')
plt.xlabel('Value')
plt.ylabel('Density')
```

```python
# Plotting CDF (Cumulative Distribution Function)
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], cumulative=True, stat='density', color='green', bins=30)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Density')
```
